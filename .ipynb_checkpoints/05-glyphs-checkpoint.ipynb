````python
def contouru (path) : # OK for a tldraw with unique record of type draw, utilise perfect_freehand
    data = load_tldraw(path)
    draw_record = next(record for record in data['records'] if record['typeName'] == 'shape' and record.get('type') == 'draw')
    x = [point['x'] for segment in draw_record['props']['segments'] for point in segment['points']]
    y = [point['y'] for segment in draw_record['props']['segments'] for point in segment['points']]    
    # Use perfect_freehand to get the stroke
    input_points = [InputPoint(x=xi, y=yi) for xi, yi in zip(x, y)]
    stroke = get_stroke(input_points)
    
    # Extract points from the stroke
    stroke_x = [point[0] for point in stroke]
    stroke_y = [point[1] for point in stroke]
    
    # Plot the skeleton points
    plt.figure(figsize=(8, 8))
    plt.plot(x, y, color=draw_record['props']['color'], linestyle='-', label='Skeleton Points')
    
    # Plot the contour
    plt.fill_between(stroke_x, stroke_y, color='lightgray', label='Contour')

    plt.title('Skeleton Points and Contour')
    
    plt.title('Skeleton Points and Countour')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.gca().invert_yaxis()
    plt.legend()
    plt.show()

contouru(Curve6path)
````

````python

def draw_glyph_variations(path):
    # Initialize empty lists to store all x and y coordinates
    all_x = []
    all_y = []

    data = load_tldraw(path)
    draw_records = [record for record in data['records'] if record['typeName'] == 'shape' and record.get('type') == 'draw']

    # Set up a colormap, 10 couleurs distinctes
    cmap = plt.get_cmap('tab10')

    # Plot all points in a single plot
    plt.figure(figsize=(8, 8))

    # Iterate over all draw_records
    for i, draw_record in enumerate(draw_records):
        # Extract x and y coordinates from each segment
        x = [point['x'] for segment in draw_record['props']['segments'] for point in segment['points']]
        y = [point['y'] for segment in draw_record['props']['segments'] for point in segment['points']]

        # Calculate the center of mass for the current glyph
        center_x = np.mean(x)
        center_y = np.mean(y)

        # Translate the coordinates to have the same center of mass
        x_translated = [coord - center_x for coord in x]
        y_translated = [coord - center_y for coord in y]

        # Extend the overall lists with the translated coordinates
        all_x.extend(x_translated)
        all_y.extend(y_translated)

        # Plot each glyph as a continuous line in a different color
        color = cmap(i / len(draw_records))
        plt.plot(x_translated, y_translated, color=color, label=f'Glyph {i+1}')

    plt.title('Combined Glyph Variations')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.gca().invert_yaxis()
    plt.show()

# Example usage
Glyphs1path = "/Users/ethanorel/tldraw-project/docs/Glyphs1.tldr"
draw_glyph_variations(Glyphs1path)
````

```{exercise} Interpolation and Resampling
 1. Associate to any glyph represented by its coordinate a path $t \in [0, 1] \mapsto \mathbb{R}^2$ (as a Python function), using some "smooth" interpolation.
 
    **Hint:** use [`scipy.interpolate.interp1d`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp1d.html).
 
 2. Sample it uniformly to get exactly 1000 points.

```


````python
# Question 1, use scipy.interpolate.interp1d.
def interpolate_smoothly(x, y):
    # Créer des valeurs de paramètres de 0 à 1
    t = np.linspace(0, 1, len(x))
    
    # Utiliser l'interpolation cubique pour les coordonnées x et y
    interp_func_x = interp1d(t, x, kind='cubic')
    interp_func_y = interp1d(t, y, kind='cubic')
    
    # Retourner les fonctions d'interpolation
    return interp_func_x, interp_func_y
````

````python
# Question 2 : 

def sample_uniformly(interp_func_x, interp_func_y, num_points=1000):
    # Créer des valeurs de paramètres uniformément espacées pour l'échantillonnage
    uniform_t = np.linspace(0, 1, num_points)
    
    # Interpoler les coordonnées x et y aux valeurs de paramètres uniformes
    sampled_x = interp_func_x(uniform_t)
    sampled_y = interp_func_y(uniform_t)
    
    return sampled_x, sampled_y
````

```{exercise} Mean Glyph
 1. Use the set of resampled coordinates to compute the (resampled) "mean glyph" that averages the features of all glyphs. 
 2. Draw this glyph !
```

![](images/glyphs-mean.jpg)

````python
# Question 1

# Fonction pour extraire les glyphes du chemin et les échantillonner
def extract_and_sample_glyphs(path, num_points=1000):
    data = load_tldraw(path)
    draw_records = [record for record in data['records'] if record['typeName'] == 'shape' and record.get('type') == 'draw']

    sampled_glyphs_list = []

    for draw_record in draw_records:
        x = [point['x'] for segment in draw_record['props']['segments'] for point in segment['points']]
        y = [point['y'] for segment in draw_record['props']['segments'] for point in segment['points']]

        fx, fy = interpolate_smoothly(x, y)
        sampled_x, sampled_y = sample_uniformly(fx, fy, num_points)

        sampled_glyphs_list.append(np.array([sampled_x, sampled_y]))

    return sampled_glyphs_list

# Fonction pour calculer le glyphe moyen
def compute_mean_glyph(sampled_glyphs):
    mean_x = np.mean([glyph[0] for glyph in sampled_glyphs], axis=0)
    mean_y = np.mean([glyph[1] for glyph in sampled_glyphs], axis=0)
    return mean_x, mean_y

# Question 2 : 

sampled_glyphs_list = extract_and_sample_glyphs(Glyphs1path)
mean_glyph = compute_mean_glyph(sampled_glyphs_list)

# Fonction pour tracer le glyphe moyen
def plot_mean_glyph(mean_glyph):
    plt.figure(figsize=(8, 8))
    plt.plot(mean_glyph[0], mean_glyph[1], color='black', linestyle='-', label='Mean Glyph')
    plt.title('Mean Glyph')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.gca().invert_yaxis()
    plt.legend()
    plt.show()

# Exemple d'utilisation
plot_mean_glyph(mean_glyph)
````

```{exercise} Glyph Similarity

 1. Come up with a scalar measure of the "dissimilarity" between two resampled glyphs (2 arrays of 1000 points), which is equal to 0 when both glyphs are identical and gets larger as the glyphs become more different.

 2. Compute the dissimilarity between each glyph and the mean glyph and compute the maximal dissimilarity for our collection of small variants.
```

Question 1

Pour mesurer la "dissimilarity" entre deux glyphes, nous pouvons utiliser la distance euclidienne entre les points correspondants des deux glyphes. Plus précisément, nous pouvons définir la dissimilarité comme la somme des carrés des distances euclidiennes entre les points correspondants des deux glyphes (normalisée par la longueur du tableau) Cette mesure sera égale à zéro lorsque les glyphes sont identiques et augmentera à mesure qu'ils deviennent plus différents.

````python
# Question 1

def dissimilarity(glyph1, glyph2):
    # Unpack les coordonnées : un glyph = un tuple de deux arrays, un contenant toutes les abscisses, l'autre toutes les ordonnées
    x1, y1 = glyph1
    x2, y2 = glyph2
    
    # Calcul de la dissimilarity
    dist_squared = np.sum((x1 - x2) ** 2 + (y1 - y2) ** 2)
    dist = np.sqrt(dist_squared)
    normalized_dist = dist / np.sqrt(len(x1))
    return normalized_dist
````

````python
# Question 2 : 

# Calculate dissimilarity between each glyph and the mean glyph
dissimilarities = [dissimilarity(glyph, mean_glyph) for glyph in sampled_glyphs_list]

max_dist = np.max(dissimilarities)

max_dist

# Fixons un seuil de 40 (max_dist) pour l'écart type acceptable
````


```{exercise} Glyph Classification

 1. Create a document with several glyphs similar to the reference one and several ones wich are dissimilar. Save it as `docs/glyphs-misc.tldr".

 2. Generate automatically a new tldraw document `docs/glyphs-classified.tldr` with the same set of glyphs, but such that the glyphs are green when they are similar to the reference one and red if they are dissimilar.
```


![](images/glyphs-misc.png)

![](images/glyphs-misc-checked.png)

````python
def check_glyphs(path, pathref):
    # Charger le document tldraw du deuxième chemin
    data = load_tldraw(path)
    sampled_glyphs_list = extract_and_sample_glyphs(path)
    sampled_glyphs_list_ref = extract_and_sample_glyphs(pathref)
    mean_glyph = compute_mean_glyph(sampled_glyphs_list_ref)
    dissimilarities = [dissimilarity(glyph, mean_glyph) for glyph in sampled_glyphs_list_ref]
    threshold = np.max(dissimilarities)

    # Initialiser le compteur de boucles inutiles à 0
    c = 0

    # Parcourir tous les enregistrements pour mettre à jour la couleur
    for i, record in enumerate(data['records']):
        # Vérifier s'il s'agit d'une forme de dessin
        if record['typeName'] == 'shape' and record.get('type') == 'draw':
            # Extraire les coordonnées du glyph
            glyph = sampled_glyphs_list[i - c]

            # Calculer la dissimilitude
            dist = dissimilarity(glyph, mean_glyph)

            # Déterminer la couleur en fonction de la dissimilitude
            color = 'green' if dist <= threshold else 'red'

            # Mettre à jour la couleur dans les propriétés de la forme
            record['props']['color'] = color

        else:
            # Incrémenter le compteur de boucles inutiles
            c += 1

    # Enregistrer le nouveau document tldraw avec les glyphes mis à jour
    with open('docs/glyphs-classified.tldr', 'w') as tldraw_file:
        json.dump(data, tldraw_file, indent=2)

check_glyphs(Glyphs3path,Glyphs1path)
````
