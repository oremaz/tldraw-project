````python
import json

def is_json_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            json.load(file)
        return True
    except (json.JSONDecodeError, FileNotFoundError):
        return False
        
# In Python, try and except are used for exception handling, allowing you to gracefully manage and respond to errors that might occur during the execution of your code.

# try block:

# The code inside the try block is the segment of code where you anticipate that an exception might occur. It's the part of your code that you want to "try" or attempt to execute.
# except block:

# The code inside the except block is executed when an exception occurs within the corresponding try block. It provides a way to handle specific types of exceptions or to execute code that should run in the event of an error.


print (is_json_file (hello1_path))
print ("Cela se voyait à la structure de dictionnaire qui apparaît lorsqu'on ouvre le fichier")
````

````python
import json

def load_tldraw(file_name):
    with open(file_name, 'r', encoding='utf-8') as file:
        tldraw = json.load(file)
    return tldraw

hello = load_tldraw (hello1_path)
hello
````

# Records & Shapes

```{exercise} Structure
:label:

 1. Can you locate a list of "records" inside `doc`? Name that list `records`.
```

````python
records = hello["records"]
assert isinstance (records, list)
print(records)
# This code locates "recorts" inside doc, then assert "records" is a list.
````

````python
type_names = [record["type"] for record in records if "type" in record]
print("Record Types:")
print(type_names)
````

````python
print ("Single shape : its type is 'draw'")
````

````python
draw_records = [record for record in records if record.get("type") == "draw"]
from IPython.display import display, Markdown
import os
output_directory = "docs"
output_file_path = os.path.join(output_directory, "hello_text.tldr")

os.makedirs(output_directory, exist_ok=True)

with open(output_file_path, "w", encoding="utf-8") as output_file:
    json.dump({"records": draw_records}, output_file, ensure_ascii=False, indent=2)

# Afficher le contenu du fichier dans le notebook
with open(output_file_path, "r", encoding="utf-8") as input_file:
    content = input_file.read()
    display(Markdown(f"**Contenu du fichier 'hello_text.tldr':**\n\n{content}"))
````
